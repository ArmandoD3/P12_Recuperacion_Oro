### 1.2. Comprueba que el c√°lculo de la recuperaci√≥n sea correcto.

```python
# Selecciona las columnas relevantes para el c√°lculo de la recuperaci√≥n y la recuperaci√≥n real
cols_for_recovery_calc = [
    'rougher.output.concentrate_au',
    'rougher.input.feed_au',
    'rougher.output.tail_au',
    'rougher.output.recovery'
]

# Crea un DataFrame temporal con solo estas columnas y eliminar filas con cualquier NaN en ellas
# Esto asegura que todas las entradas para el c√°lculo y la comparaci√≥n sean v√°lidas
df_recovery_check = df_train[cols_for_recovery_calc].dropna().copy()

# Renombra para mayor claridad en el c√°lculo
C = df_recovery_check['rougher.output.concentrate_au']
F = df_recovery_check['rougher.input.feed_au']
T = df_recovery_check['rougher.output.tail_au']

# Calcula el denominador, asegur√°ndose de que no sea cero para evitar divisiones por cero
denominator = F * (C - T)

# Crea una serie para la recuperaci√≥n calculada
calculated_recovery = pd.Series(0.0, index=df_recovery_check.index) # Inicializa con ceros

# Realiza el c√°lculo solo donde el denominador no es cero
# Esto es crucial: si el denominador es 0, la recuperaci√≥n es 0 (no hay material para recuperar o no hay cambio)
non_zero_denominator_mask = (denominator != 0)

calculated_recovery[non_zero_denominator_mask] = (C[non_zero_denominator_mask] * (F[non_zero_denominator_mask] - T[non_zero_denominator_mask])) / \
                                                  denominator[non_zero_denominator_mask] * 100 # Multiplica por 100 para porcentaje

# Si F es 0 y C-T no es 0, el denominador es 0. En ese caso la recuperaci√≥n es 0.
# Si F no es 0 pero C-T es 0, el denominador es 0. En ese caso, significa que la concentraci√≥n en el concentrado es igual a la cola, por lo que no hay "recuperaci√≥n" real.
# La f√≥rmula original asume que la recuperaci√≥n es un porcentaje, por lo que multiplicamos por 100.

# Recuperaci√≥n real del dataset
actual_recovery = df_recovery_check['rougher.output.recovery']

# Calcula el EAM
mae = mean_absolute_error(actual_recovery, calculated_recovery)

print(f"\nError Absoluto Medio (EAM) entre la recuperaci√≥n calculada y la real para rougher.output.recovery (ajustado): {mae:.4f}")
```

Esto confirma que la f√≥rmula de recuperaci√≥n es correcta y que nuestros c√°lculos coinciden perfectamente con los valores proporcionados en el dataset de entrenamiento para rougher.output.recovery

#### 1.4.3 Eliminar columnas que no est√°n en df_test de df_train

```python
# Obtenemos las columnas del df_test
test_cols = set(df_test.columns)

# Las columnas objetivo siempre deben ser parte del df_train (para entrenamiento)
# y las necesitamos para comparar con las predicciones del modelo en df_test despu√©s.
# Vamos a extraer los objetivos de df_full ya que df_test no los tiene.
# Primero, asegur√©monos de que df_full tambi√©n tiene su √≠ndice 'date' y est√° ordenado.
df_full.sort_index(inplace=True)

# Unimos las columnas objetivo de df_full a df_test bas√°ndose en el √≠ndice de fecha
# Esto es crucial para poder calcular el sMAPE en el conjunto de prueba al final.
# Unimos por √≠ndice 'date' asegura que los objetivos se asignen a las filas correctas.
df_test = df_test.merge(df_full[['rougher.output.recovery', 'final.output.recovery']], 
                        left_index=True, right_index=True, how='left')

# Ahora, eliminamos las columnas de df_train que no est√°n en el df_test (excluyendo los objetivos que acabamos de 
# a√±adir a df_test para alineaci√≥n de evaluaciones)
# Las columnas objetivo que necesitamos para el entrenamiento son 'rougher.output.recovery' y 'final.output.recovery'.
# Estas son las que usaremos para entrenar en df_train, pero solo las predicciones se har√°n sobre df_test.
# Por lo tanto, no se deben eliminar de df_train.
# Para las caracter√≠sticas predictoras, S√ìLO usaremos las que est√°n en df_test.

# Identificamos las caracter√≠sticas (features) que debemos mantener en df_train
# Estas son las columnas que est√°n en df_test (excepto 'date', que ya es el √≠ndice)
features_to_keep_in_train = list(test_cols)

# A√±adimos las columnas objetivo a la lista de columnas a mantener en df_train
# Esto es esencial porque df_train es nuestro conjunto de entrenamiento completo con targets
features_to_keep_in_train.append('rougher.output.recovery')
features_to_keep_in_train.append('final.output.recovery')

# Eliminamos cualquier duplicado y nos aseguramos de que sean columnas existentes en df_train
features_to_keep_in_train = list(set(features_to_keep_in_train).intersection(df_train.columns))

# Filtramos df_train para que solo contenga estas caracter√≠sticas y los objetivos
df_train = df_train[features_to_keep_in_train]

print(f"\nN√∫mero de columnas en df_train despu√©s de la limpieza: {df_train.shape[1]}")
print(f"N√∫mero de columnas en df_test despu√©s de a√±adir objetivos: {df_test.shape[1]}")

# Verificamos si a√∫n hay NaNs despu√©s del ffill/bfill (deber√≠a ser muy pocos o ninguno)
print("\nConteo de NaNs despu√©s del preprocesamiento en df_train:")
print(df_train.isnull().sum().sum()) # Suma total de NaNs en df_train

print("\nConteo de NaNs despu√©s del preprocesamiento en df_test:")
print(df_test.isnull().sum().sum()) # Suma total de NaNs en df_test
```

### 2.1. Observa c√≥mo cambia la concentraci√≥n de metales (Au, Ag, Pb) en funci√≥n de la etapa de purificaci√≥n.

```python
# Utilizaremos el DataFrame df_full para este an√°lisis, ya que contiene todas las etapas y sus concentraciones.
# Definimos las columnas de concentraci√≥n para cada metal en cada etapa
metals = ['au', 'ag', 'pb']
stages = {
    'Materia Prima': 'rougher.input.feed_',
    'Concentrado Rougher': 'rougher.output.concentrate_',
    'Limpieza Primaria': 'primary_cleaner.output.concentrate_',
    'Concentrado Final': 'final.output.concentrate_'
}

# Creamos un DataFrame para almacenar las concentraciones promedio
concentration_data = pd.DataFrame(columns=['Etapa', 'Metal', 'Concentraci√≥n Promedio'])

for metal in metals:
    for stage_name, col_prefix in stages.items():
        col_name = col_prefix + metal
        if col_name in df_full.columns: # Asegurarse de que la columna exista
            avg_concentration = df_full[col_name].mean()
            concentration_data = pd.concat([concentration_data, pd.DataFrame([{'Etapa': stage_name, 'Metal': metal, 'Concentraci√≥n Promedio': avg_concentration}])], ignore_index=True)

# Visualiza los resultados
plt.figure(figsize=(12, 7))
sns.barplot(x='Etapa', y='Concentraci√≥n Promedio', hue='Metal', data=concentration_data, palette='viridis')
plt.title('Cambio en la concentraci√≥n promedio de metales por etapa de purificaci√≥n')
plt.xlabel('Etapa de Purificaci√≥n')
plt.ylabel('Concentraci√≥n Promedio')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Metal')
plt.show()

print("\nConcentraciones promedio de metales por etapa de purificaci√≥n:")
print(concentration_data)
```

### 2.3. Considera las concentraciones totales de todas las sustancias en las diferentes etapas: materia prima, concentrado rougher y concentrado final.

```python
# Definimos las columnas para cada etapa y metal/sustancia
stages_concentrations = {
    'Materia Prima (Alimentaci√≥n)': ['rougher.input.feed_au', 'rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol'],
    'Concentrado Rougher': ['rougher.output.concentrate_au', 'rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_sol'],
    'Concentrado Final': ['final.output.concentrate_au', 'final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_sol']
}

# Calcula la concentraci√≥n total para cada etapa
for stage_name, cols in stages_concentrations.items():
    # las columnas existan antes de sumarlas?
    existing_cols = [col for col in cols if col in df_full.columns]
    df_full[f'total_concentration_{stage_name.replace(" ", "_").lower()}'] = df_full[existing_cols].sum(axis=1)
    print(f"Calculada la concentraci√≥n total para: {stage_name}")

# Visualizamos las distribuciones de las concentraciones totales
plt.figure(figsize=(15, 6))

for i, (stage_name, _) in enumerate(stages_concentrations.items()):
    col_name = f'total_concentration_{stage_name.replace(" ", "_").lower()}'
    plt.subplot(1, len(stages_concentrations), i + 1)
    sns.histplot(df_full[col_name], kde=True, bins=50)
    plt.title(f'Distribuci√≥n de Concentraci√≥n Total en {stage_name}')
    plt.xlabel('Concentraci√≥n Total')
    plt.ylabel('Frecuencia')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Analiza valores anormales (cercanos a cero)
print("\n--- An√°lisis de valores anormales (concentraci√≥n total cercana a cero) ---")
for stage_name, _ in stages_concentrations.items():
    col_name = f'total_concentration_{stage_name.replace(" ", "_").lower()}'
    # Contar cu√°ntas filas tienen una concentraci√≥n total muy baja (cercana a cero)
    # Por ejemplo, < 1.0 o incluso menos si la distribuci√≥n muestra un pico en 0.
    # Usaremos un umbral de 0.1 para identificar valores "casi cero"
    abnormal_count = df_full[df_full[col_name] < 0.1].shape[0]
    total_count = df_full.shape[0]
    print(f"Etapa '{stage_name}': {abnormal_count} filas con concentraci√≥n total < 0.1 (de {total_count} total).")

    # Si hay valores anormales, vemos sus estad√≠sticas o algunos ejemplos
    if abnormal_count > 0:
        print(f"Estad√≠sticas para valores anormales en '{stage_name}':")
        print(df_full[df_full[col_name] < 0.1][col_name].describe())
```

Podemos ver unos picos anormales al inicio de cada grafica por lo que los eliminaremos para evitar errores en la prediccion,
utilizamos df_full para identificar los √≠ndices a eliminar para asegurar la consistencia, y luego aplicamos esa eliminaci√≥n a df_train y df_test individualmente.

# 3. Contruye el modelo

### 3.1. Escribe una funci√≥n para calcular el valor final de sMAPE.

```python
# Definir las funciones smape y final_smape 
def smape(y_true, y_pred):
    numerator = np.abs(y_true - y_pred)
    denominator = (np.abs(y_true) + np.abs(y_pred)) / 2
    smape_value_per_sample = np.where(denominator == 0, 0, numerator / denominator)
    return np.mean(smape_value_per_sample) * 100

def final_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher + 0.75 * smape_final
```

<b>√âxito</b> - Has definido correctamente la m√©trica sMAPE, una elecci√≥n adecuada para problemas donde los errores relativos son importantes. Tambi√©n es muy positivo que encapsulaste el c√°lculo ponderado en final_smape, lo cual te ayuda a mantener el c√≥digo limpio y reutilizable. ‚úÖ

### 3.2. Entrena diferentes modelos

#### 3.2.1: Preparaci√≥n de X e Y

```python
# Obtenemos las columnas de caracter√≠sticas a usar (las que est√°n en df_test como features)
target_cols = ['rougher.output.recovery', 'final.output.recovery']
features_cols = [col for col in df_test.columns if col not in target_cols]

# Asegurar que df_train solo contenga las caracter√≠sticas y objetivos relevantes
# Esto es vital para que las columnas de X_train y X_test coincidan.
required_cols_in_train = features_cols + target_cols
df_train = df_train[[col for col in required_cols_in_train if col in df_train.columns]]

# Dividimos los datos de entrenamiento en caracter√≠sticas y objetivos
X_train = df_train[features_cols]
y_train_rougher = df_train['rougher.output.recovery']
y_train_final = df_train['final.output.recovery']

# Preparar X_test (solo caracter√≠sticas)
X_test = df_test[features_cols]
```

#### 3.2.2 Configuraci√≥n de Scorer y Modelos, Entrenamiento y Evaluaci√≥n CV

```python
# Definimos los scorers sMAPE personalizados
smape_scorer_rougher = make_scorer(smape, greater_is_better=False)
smape_scorer_final = make_scorer(smape, greater_is_better=False)

# Modelos y sus par√°metros para GridSearchCV
models = {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {}
    },
    'DecisionTreeRegressor': {
        'model': DecisionTreeRegressor(random_state=123),
        'params': {
            'max_depth': [5, 10, 15]
        }
    },
    'RandomForestRegressor': {
        'model': RandomForestRegressor(random_state=123),
        'params': {
            'n_estimators': [10, 30, 50],
            'max_depth': [5, 10]
        }
    }
}

best_models = {}
best_scores = {}

# Entrenamiento y evaluaci√≥n para la Recuperaci√≥n Rougher
print("--- Entrenando y evaluando modelos para rougher.output.recovery ---")
for model_name, config in models.items():
    print(f"\nProcesando {model_name} para Rougher Recovery...")
    grid_search = GridSearchCV(
        estimator=config['model'],
        param_grid=config['params'],
        scoring=smape_scorer_rougher,
        cv=5,
        n_jobs=-1,
        verbose=1
    )
    grid_search.fit(X_train, y_train_rougher)

    best_models[f'rougher_{model_name}'] = grid_search.best_estimator_
    best_scores[f'rougher_{model_name}_smape'] = -grid_search.best_score_
    print(f"Mejores par√°metros para {model_name} (Rougher): {grid_search.best_params_}")
    print(f"Mejor sMAPE CV para {model_name} (Rougher): {best_scores[f'rougher_{model_name}_smape']:.4f}")

# Entrenamiento y evaluaci√≥n para la Recuperaci√≥n Final
print("\n--- Entrenando y evaluando modelos para final.output.recovery ---")
for model_name, config in models.items():
    print(f"\nProcesando {model_name} para Final Recovery...")
    grid_search = GridSearchCV(
        estimator=config['model'],
        param_grid=config['params'],
        scoring=smape_scorer_final,
        cv=5,
        n_jobs=-1,
        verbose=1
    )
    grid_search.fit(X_train, y_train_final)

    best_models[f'final_{model_name}'] = grid_search.best_estimator_
    best_scores[f'final_{model_name}_smape'] = -grid_search.best_score_
    print(f"Mejores par√°metros para {model_name} (Final): {grid_search.best_params_}")
    print(f"Mejor sMAPE CV para {model_name} (Final): {best_scores[f'final_{model_name}_smape']:.4f}")

print("\n--- Resumen de los mejores sMAPE CV por modelo y etapa ---")
for key, value in best_scores.items():
    print(f"{key}: {value:.4f}")

# Selecci√≥n del Mejor Modelo Combinado
print("\n--- Calculando sMAPE final ponderado para las mejores combinaciones ---")
combined_smape_scores = {}

for model_name in models.keys():
    smape_rougher = best_scores[f'rougher_{model_name}_smape']
    smape_final = best_scores[f'final_{model_name}_smape']
    combined_smape = final_smape(smape_rougher, smape_final)
    combined_smape_scores[model_name] = combined_smape
    print(f"sMAPE Final Ponderado para {model_name}: {combined_smape:.4f}")

# Encontrar el mejor modelo general
best_overall_model_name = min(combined_smape_scores, key=combined_smape_scores.get)
print(f"\nEl mejor modelo general (con el sMAPE final ponderado m√°s bajo) es: {best_overall_model_name}")

# Almacenar los modelos seleccionados para la prueba final
best_rougher_model = best_models[f'rougher_{best_overall_model_name}']
best_final_model = best_models[f'final_{best_overall_model_name}']

print(f"\nLos modelos seleccionados para la prueba final son:")
print(f"  - Rougher: {best_rougher_model.__class__.__name__} con par√°metros {best_rougher_model.get_params()}")
print(f"  - Final: {best_final_model.__class__.__name__} con par√°metros {best_final_model.get_params()}")
```

<b>√âxito</b> - Excelente enfoque al preparar los datos, definir make_scorer personalizado, y usar GridSearchCV con cv=5 para cada modelo y objetivo. Adem√°s, evaluaste todos los modelos de forma consistente y seleccionaste el mejor bas√°ndote en un criterio bien definido (sMAPE final ponderado). ¬°Una implementaci√≥n de pipeline muy profesional!

### 3.3. Prueba el mejor modelo utilizando la muestra de prueba.

```python
# Re-entrenar los modelos seleccionados con el conjunto de entrenamiento completo
# (Esto ya lo hace GridSearchCV internamente al obtener .best_estimator_, pero lo explicitamos aqu√≠)
best_rougher_model.fit(X_train, y_train_rougher)
best_final_model.fit(X_train, y_train_final)

# Realizar predicciones en el conjunto de prueba
predictions_rougher = best_rougher_model.predict(X_test)
predictions_final = best_final_model.predict(X_test)

# Obtenemos los valores reales de recuperaci√≥n del conjunto de prueba
y_test_rougher = df_test['rougher.output.recovery']
y_test_final = df_test['final.output.recovery']

# Calculamos sMAPE para cada etapa en el conjunto de prueba
smape_rougher_test = smape(y_test_rougher, predictions_rougher)
smape_final_test = smape(y_test_final, predictions_final)

print(f"\nsMAPE para rougher.output.recovery en la muestra de prueba: {smape_rougher_test:.4f}")
print(f"sMAPE para final.output.recovery en la muestra de prueba: {smape_final_test:.4f}")

# Calculamos el sMAPE final ponderado
final_smape_test = final_smape(smape_rougher_test, smape_final_test)

print(f"\n--- sMAPE FINAL PONDERADO en la MUESTRA DE PRUEBA: {final_smape_test:.4f} ---")
```
